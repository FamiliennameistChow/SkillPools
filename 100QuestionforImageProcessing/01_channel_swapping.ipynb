### 1. 通道转换
opencv中读出的图像格式为BGR格式
```
cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

````python
import cv2
import matplotlib.pyplot as plt
````

````python
img = cv2.imread("gg.jpg")
b = img[:, :, 0].copy()
g = img[:, :, 1].copy()
r = img[:, :, 2].copy()
plt.figure()
plt.subplot(121)
plt.imshow(img)
plt.title("BGR")

img[:, :, 0] = r
img[:, :, 1] = g
img[:, :, 2] = b

plt.subplot(122)
plt.imshow(img)
plt.title("RGB")

plt.show()
````
