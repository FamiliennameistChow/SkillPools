最近邻插值在图像放大时补充的像素取最临近的像素的值。由于方法简单，所以处理速度很快，但是放大图像画质劣化明显



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("gg.jpg")
img_origin = img.copy()

(H, W, C) = img.shape

# 放大倍数
ratio = 1.5

Height = int(H * ratio)
Width = int(W * ratio)

img_out = np.zeros((Height, Width, C), dtype=np.uint8)

for height in range(Height):
    for width in range(Width):
        for c in range(C):
            img_out[height, width, c] = img[int(height/ratio), int(width/ratio), c]

img_out = img_out.astype(np.uint8)


# 方便plt显示
img_origin = cv2.cvtColor(img_origin, cv2.COLOR_BGR2RGB)
img_out = cv2.cvtColor(img_out, cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(img_origin)

plt.figure()
plt.imshow(img_out)

# plt.figure()
# plt.subplot(121)
# plt.hist(img.ravel(), bins=255, rwidth=1, range=(0, 255))
# plt.subplot(122)
# plt.imshow(img_origin)
# plt.title("img_origin")

# plt.figure()
# plt.subplot(121)
# plt.hist(img_out.ravel(), bins=255, rwidth=1, range=(0, 255))
# plt.subplot(122)
# plt.imshow(img_out)
# plt.title("img_out")

# # plt.subplots_adjust(top=1.2, bottom=0.08, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

# plt.show()
```
