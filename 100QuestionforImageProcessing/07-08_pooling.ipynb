### Q7 平均池化
将图片按照固定大小网格分割，网格内的像素值取网格内所有像素的平均值。我们将这种把图片使用均等大小网格分割，
并求网格内代表值的操作称为池化（Pooling）。池化操作是卷积神经网络（Convolutional Neural Network）中重要的图像处理方式。
平均池化按照下式定义：
$$v = 1/|R| * Sum_{i in R} v_i$$

### Q8 最大池化
网格内的值不取平均值，而是取网格内的最大值进行池化操作。

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("gg.jpg")
img_origin = img.copy()

H, W, C = img.shape
print(H, W, C)

kernel_size = 8
POOL_MODE = 'max_pooling'  # "average_pooling"

hs = int(H / kernel_size)
ws = int(W / kernel_size)
for h in range(hs):
    for w in range(ws):
        for c in range(C):
            if POOL_MODE == 'average_pooling':
                img[h*kernel_size:kernel_size*(h+1), w*kernel_size:kernel_size*(w+1), c] \
                = np.mean(img[h*kernel_size:kernel_size*(h+1), w*kernel_size:kernel_size*(w+1), c]).astype(np.int)
            if POOL_MODE == 'max_pooling':
                img[h*kernel_size:kernel_size*(h+1), w*kernel_size:kernel_size*(w+1), c] \
                = np.max(img[h*kernel_size:kernel_size*(h+1), w*kernel_size:kernel_size*(w+1), c])
                

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_origin = cv2.cvtColor(img_origin, cv2.COLOR_BGR2RGB)            


plt.figure()
plt.subplot(121)
plt.imshow(img)
plt.title(POOL_MODE)
plt.subplot(122)
plt.imshow(img_origin)
plt.title("img_origin")
plt.show()
```
