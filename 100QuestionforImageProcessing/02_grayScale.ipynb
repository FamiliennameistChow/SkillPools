### 图像灰度化操作
灰度是一种图像亮度的表示方法，通过下式计算：
> Y = 0.2126 R + 0.7152 G + 0.0722 B

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("gg.jpg")
# img = cv2.imread("gg.jpg").astype(np.float)
b = img[:, :, 0].copy()
g = img[:, :, 1].copy()
r = img[:, : ,2].copy()

plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("rgb")

gray = 0.2126*r + 0.7152*g + 0.0722*b
img_gray = gray.astype(np.uint8)

plt.subplot(1, 2, 2)
plt.imshow(img_gray,  cmap='gray')
plt.title("gray")

plt.show()
```
