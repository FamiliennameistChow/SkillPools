### 二值化操作
二值化是将图像使用黑和白两种值表示的方法

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("gg.jpg")
b = img[:, :, 0].copy()
g = img[:, :, 1].copy()
r = img[:, :, 2].copy()

plt.figure()
plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title("rgb")

# GrayScale
img_gray = (0.2126 * r + 0.7152 * g + 0.0722 * b).astype(np.uint8)


plt.subplot(1, 3, 2)
plt.imshow(img_gray, cmap='gray')
plt.title("img_gray")
```

```python
# binarization
thersh = 128
img_gray[img_gray >= thersh] = 255
img_gray[img_gray < thersh] = 0


plt.subplot(1, 3, 3)
plt.imshow(img_gray, cmap='gray')
plt.title("img_bin")

plt.show()
```
